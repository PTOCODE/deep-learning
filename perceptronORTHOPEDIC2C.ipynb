Perceptron on the binary classification problem of Biomechanical
features of orthopedic patients.

```
from sklearn.linear_model import Perceptron
```

```
dataset=pd.read_csv('/content/drive/My Drive/datasets/column_2C_weka.csv')
```

```
X=dataset.iloc[:,:-1].values
y=dataset.iloc[:,6].values
```

```
print (dataset.isnull().values.any())
```

CATEGORICAL

```
from sklearn.preprocessing  import LabelEncoder,OneHotEncoder
labelencoder_y=LabelEncoder()
y=labelencoder_y.fit_transform(y)
```

```
from sklearn.model_selection import train_test_split
# While splitting, make an unbiased splitting
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y,
random_state=42)
# Create an object (model) for Perceptron.
```

```

perceptronClassifier = Perceptron()
# Train the perceptron using 80% training set
perceptronClassifier.fit(X_train, y_train)

score = perceptronClassifier.score(X_test, y_test)

print(score)
```
