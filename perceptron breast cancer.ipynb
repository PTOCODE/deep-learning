```
import numpy as np
# Load the dataset from sklearn
from sklearn.datasets import load_breast_cancer
# Use the Perceptron class defined in sklearn
from sklearn.linear_model import Perceptron
# Load the dataset
cancerDataset = load_breast_cancer()
```

```
print(cancerDataset.DESCR)
# Load the attributes and target in X and y
X = cancerDataset.data
y = cancerDataset.target
# Look at the unique number of labels present in y
print(set(y))
# {0, 1}
```

```
X = cancerDataset.data
y = cancerDataset.target
# Look at the unique number of labels present in y
print(set(y))
# {0, 1}
# Split the data into 80% training and 20% testing
from sklearn.model_selection import train_test_split
# While splitting, make an unbiased splitting
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y,
random_state=42)
# Create an object (model) for Perceptron.
perceptronClassifier = Perceptron()
# Train the perceptron using 80% training set
perceptronClassifier.fit(X_train, y_train)
```

```
score = perceptronClassifier.score(X_test, y_test)
# The classification score accuracy obtained.
print(score)
# 0.9122807017543859
```
